## Save numerical data only
Slicing with .loc and .iloc


```python
x1 = data1.loc[:4,] 
print(x1, "\n") 
```

```python
dt1 = data1[data1.columns[1:]]
dt1.index = data1.iloc[:,0]

dt2 = data2[data2.columns[1:]]
dt2.index = data2.iloc[:,0]
```

```python
list(dt1.columns.values)
```

# Analysis
## Create Linkage Matrix for Clustering

```python
Cl1 = linkage(dt1, 'ward')
Cl2 = linkage(dt2, 'ward')
```

## Plot Dendrogram of Clusters 1

```python
plt.figure(figsize = (25, 10))
plt.title('Cluster with unicorn, internet security, data mining, brexit, fifa')
plt.ylabel('Distance')
dendrogram(
    Cl1,
    labels = dt1.index,
    leaf_rotation = 90.,
    leaf_font_size = 10.,
)
plt.show()
```

## Plot Dendrogram of Clusters 2


```python
plt.figure(figsize = (25, 10))
plt.title('Cluster with startup, ceo, interpreneur, data science')
plt.ylabel('Distance')
dendrogram(
    Cl2,
    labels = dt2.index,
    leaf_rotation = 90.,
    leaf_font_size = 10.,
)
plt.show()
```
