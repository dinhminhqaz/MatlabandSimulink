### Damped simple harmonic motion equation



```python
% matplotlib inline
import numpy as np # import numpy library as np
import matplotlib.pyplot as plt # import pyplot library as plt
plt.style.use('ggplot') # use "ggplot" style for graphs
%pylab inline

from scipy.integrate import odeint
```

```python
def dS_dt(S, t):
    # Here S is a vector such that y=S[0] and v=S[1]. This function should return [y', v']
    return [S[1], -3*S[1] - 10*S[0] + np.cos(2*t)]
S0 = [0, 0]
ts = np.linspace(0, 10, 200)
Ss = odeint(dS_dt, S0, ts)
ys = Ss[:,0]
```

```python
plt.plot(ts,ys)
plt.xlabel("time")
plt.ylabel("Displacement")
plt.title("Damped harmonic oscillator");
```
